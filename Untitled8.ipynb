```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# Open file
url = 'https://en.wikipedia.org/w/index.php?title=List_of_countries_by_GDP_(nominal)&oldid=875043608'
response = requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
```

```python
# Find all tables on the page
all_tables = soup.find_all('table')
print('Total number of tables are {} '.format(len(all_tables)))
```

```python
# Find first table on html with class wikitable
data_table = soup.find('table',{'class': 'wikitable'})
print(type(data_table))
```

```python
#Find first table count of sources
sources = data_table.tbody.find_all('tr', recursive=False) [0]
sources_list = [td  for td in sources.find_all('td')]
print(len(sources_list))
```

```python
# Extract data from secong rows
data = data_table.tbody.find_all('tr', recursive = False)[1].find_all('td', recursive=False)
```

```python
# Find all with td tag
data_tables = []
for td in data:
    data_tables.append(td.find_all('table'))
```

```python
# Find length of data tables
len(data_tables)
```

```python
# Extract sources from html
source_names = [source.findAll('a')[0].getText() for source in sources_list]
print(source_names)
```

```python
#From wikitable 1 pull headers
wikitables = soup.find_all('table', class_= 'wikitable')

for i, table in enumerate(wikitables[1:2], start=1):
    for row in table.find_all('tr'):
        th_cells = row.find_all('th')
        if th_cells:
            headers = [cell.get_text(strip=True) for cell in th_cells]
            print(headers)
```

```python
# From wikitables 1 pull rows
rows= wikitables[1].find_all('tr')[1:]
data = [[td.get_text(strip=True) for td in row.find_all('td')] for row in rows if row.find_all('td')]
```

```python
# Create data frame from headers and rows of wikitable 1
df1=pd.DataFrame(data=data, columns=headers)
df1.head()
```

```python
# Filter the data to remove outliers
df1 = df[(df['visit'] <= 2900) & (df['visit'] >= 100)]
# Print size of dataframe after removing outliers
print ('After getting rid of outliers the new size of the data is - {}'.format(*df1.shape))
```

```python
# Connect to SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
# Create a table called 'contacts' with 6 columns
cursor.execute('''
CREATE TABLE contacts (
    Name TEXT,
    Address TEXT, 
    City TEXT,
    State Text,
    Zip Text, 
    PhoneNumber TEXT
)
''')
```
